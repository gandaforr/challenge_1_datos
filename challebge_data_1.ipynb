#importar tienda

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# ingreso total por cada tienda
En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
# trato de valores nulos
  #sale 0
    #tienda["Precio"].isnull().sum()
    #tienda2["Precio"].isnull().sum()
    #tienda3["Precio"].isnull().sum()
    #tienda4["Precio"].isnull().sum()
    #no hay valores nulos que tratar
#suma de los valores
#Tiendas = [tienda,tienda2,tienda3,tienda4]
''' it1 = tienda["Precio"].sum()
it2 = tienda2["Precio"].sum()
it3 = tienda3["Precio"].sum()
it4 = tienda4["Precio"].sum() '''

Tiendas = [tienda,tienda2,tienda3,tienda4]
ingresos_total = []
for t in Tiendas:
  juan = t["Precio"].sum()
  ingresos_total.append(juan)
#print(it1,it2,it3,it4)
print(ingresos_total)
```

# Ventas por categoría
En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
# agrupar por categoria
#agrupar por categoria y ordenar
''' vct = tienda.groupby('Categoría del Producto')['Cantidad de cuotas'].sum(numeric_only=True)
vct.sort_values(ascending=False)

vct2 = tienda2.groupby('Categoría del Producto')['Cantidad de cuotas'].sum(numeric_only=True)
vct2.sort_values(ascending=False)

vct3 = tienda3.groupby('Categoría del Producto')['Cantidad de cuotas'].sum(numeric_only=True)
vct3.sort_values(ascending=False)

vct4 = tienda4.groupby('Categoría del Producto')['Cantidad de cuotas'].sum(numeric_only=True)
vct4.sort_values(ascending=False) '''
ventas_por_categoria = []
for t in Tiendas:
  vct = t.groupby('Categoría del Producto')['Cantidad de cuotas'].sum(numeric_only=True)
  #print(vct.to_dict())
  #vct_normalisado = pd.DataFrame({'Categoria del producto': ["Electrodomésticos", "Electrónicos", "Juguetes", "Libros", "Muebles"],
   #                               "Cantidad de coutas": [vct.Electrodomésticos, vct.Electrónicos, vct.Juguetes, vct.Libros, vct.Muebles]})
  #vct_ordenado = vct_normalisado.sort_values(by="Cantidad de coutas",ascending=False)
  vct_ordenado = vct.sort_values(ascending=False)
  #print(vct_ordenado)
  vct_dic = vct_ordenado.to_dict()
  #print(list(zip(vct_dic.keys(),vct_dic.values())))
  ventas_por_categoria.append(list(zip(vct_dic.keys(),vct_dic.values()))) #entiendo que el zip es lento espero que no me cueste puntos

print("---Venta por categoria----")
print(ventas_por_categoria)
#print(ventas_por_categoria[0])

# categoria mas popular de cada tienda
categoria_mas_popular = []
for ven in ventas_por_categoria:
  popular = ven[0]
  categoria_mas_popular.append(popular)

print("---categoria mas popular----")
print(categoria_mas_popular)
```

#Valoración media por tienda
En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
tienda['Calificación'].mean()
#Calificacion promedio de cada tienda
promedios = []
for t in Tiendas:
  promedios.append(t['Calificación'].mean())
print(promedios)
```

#Productos más vendidos y menos vendidos
En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
#pct = tienda.groupby('Producto')['Cantidad de cuotas'].sum(numeric_only=True)
#pct.sort_values(ascending=False)
#producto por cantidad de coutas vendidas
producto_x_coutas = []
for t in Tiendas:
  pct = t.groupby('Producto')['Cantidad de cuotas'].sum(numeric_only=True)
  pct_dic = pct.sort_values(ascending=False).to_dict()
  producto_x_coutas.append(list(zip(pct_dic.keys(),pct_dic.values())))

''' for tabla in producto_x_coutas:
  print(tabla)
  print("-------")  '''
#el producto mas vendido de cada tienda
producto_mas_vendido = []
for t in producto_x_coutas:
  producto_mas_vendido.append(t[0])


#el producto menos vendido de cada tienda
producto_menos_vendido = []
for t in producto_x_coutas:
  producto_menos_vendido.append(t[-1])

print(producto_mas_vendido)
print(producto_menos_vendido)
```

#Valor del envío promedio por tienda
En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
costo_de_envio_promedio = []
for t in Tiendas:
  costo_de_envio_promedio.append(t["Costo de envío"].mean())
print(costo_de_envio_promedio)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle("Análisis de Tiendas", fontsize=18, fontweight="bold", y=1.01)
plt.subplots_adjust(hspace=0.45, wspace=0.35)

colores = ["#4C72B0", "#DD8452", "#55A868", "#C44E52"]
df = minado
df["Tienda"]

ax1 = axes[0, 0]
bars = ax1.bar(df["Tienda"], df["Ingresos_total"] / 1e6, color=colores, edgecolor="white", linewidth=0.8)
ax1.set_title("Ingresos Totales por Tienda", fontsize=13, fontweight="bold")
ax1.set_xlabel("Tienda")
ax1.set_ylabel("Ingresos (millones $)")
ax1.yaxis.set_major_formatter(mticker.FormatStrFormatter("$%.0fM"))
for bar, val in zip(bars, df["Ingresos_total"] / 1e6):
    ax1.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 3,
             f"${val:.1f}M", ha="center", va="bottom", fontsize=9, fontweight="bold")
ax1.set_ylim(0, df["Ingresos_total"].max() / 1e6 * 1.15)
ax1.spines[["top", "right"]].set_visible(False)

ax2 = axes[0, 1]
bars2 = ax2.barh(df["Tienda"], df["Promedio de calificacion"], color=colores, edgecolor="white", linewidth=0.8)
ax2.set_title("Promedio de Calificación por Tienda", fontsize=13, fontweight="bold")
ax2.set_xlabel("Calificación promedio (sobre 5)")
ax2.set_xlim(3.8, 4.2)
for bar, val in zip(bars2, df["Promedio de calificacion"]):
    ax2.text(val + 0.003, bar.get_y() + bar.get_height() / 2,
             f"{val:.3f} ⭐", va="center", fontsize=9, fontweight="bold")
ax2.spines[["top", "right"]].set_visible(False)


ax3 = axes[1, 0]
ax3.plot(df["Tienda"], df["Costo de envío promedio"], marker="o", linewidth=2.5,
         markersize=10, color="#4C72B0", markerfacecolor="white", markeredgewidth=2.5)
ax3.fill_between(df["Tienda"], df["Costo de envío promedio"],
                 alpha=0.15, color="#4C72B0")
ax3.set_title("Costo de Envío Promedio por Tienda", fontsize=13, fontweight="bold")
ax3.set_xlabel("Tienda")
ax3.set_ylabel("Costo promedio ($)")
ax3.yaxis.set_major_formatter(mticker.FormatStrFormatter("$%.0f"))
for i, (x, y) in enumerate(zip(df["Tienda"], df["Costo de envío promedio"])):
    ax3.text(i, y + 200, f"${y:,.0f}", ha="center", fontsize=9, fontweight="bold")
ax3.set_ylim(22000, 27500)
ax3.spines[["top", "right"]].set_visible(False)

ax4 = axes[1, 1]
cat_counts = df["Categoria mas popular"].value_counts()
wedge_colors = ["#4C72B0", "#DD8452"]
wedges, texts, autotexts = ax4.pie(
    cat_counts.values,
    labels=cat_counts.index,
    autopct="%1.0f%%",
    colors=wedge_colors,
    startangle=90,
    wedgeprops={"edgecolor": "white", "linewidth": 2},
)

for t in autotexts:
    t.set_fontsize(12)
    t.set_fontweight("bold")
ax4.set_title("Distribución de Categoría\nmás Popular", fontsize=13, fontweight="bold")

plt.tight_layout()

plt.savefig('para_informe.jpg')
plt.show()

```
